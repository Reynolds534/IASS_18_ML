Changing the metric

Metric: The metric is the definition of distance used by the algorithm.

In KNN, when metric = 'distance', neighbots are weighted by the inverse of their distance.

In this case, closer neighbors of a query point will have a greater influence than neighbors which are further away.

```python
metrics = ['uniform', 'distance']
n_neighbors = 2

for m in metrics:
    KNN_plot(n_neighbors, metric = m)
```

Interpretation: When distance is used as a metric, singular points have a stronger influence.
